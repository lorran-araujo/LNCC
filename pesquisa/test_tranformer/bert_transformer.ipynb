```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Nome do modelo específico para análise de sentimentos
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Carregar o tokenizer e o modelo
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Exemplos de frases para classificação
sentences = ["I love this product!", "This was the worst experience ever."]

# Tokenização e preparação das frases para o modelo
inputs = tokenizer(sentences, return_tensors="pt", padding=True, truncation=True)

# Executar a inferência sem gradientes
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

# Converter os logits em classes
predictions = torch.argmax(logits, dim=-1)
for sentence, prediction in zip(sentences, predictions):
    label = "Positive" if prediction == 1 else "Negative"
    print(f"Sentence: '{sentence}' - Sentiment: {label}")
```
