A brincadeira vai ser: sortear uma certa quantidade de dados e calcular a soma dos resultados, e ver como a soma vai ficar distribuida

```python
import numpy as np
import matplotlib.pyplot as plt
```

Fazendo experimento e plotando histograma:

```python
num_dados = 6
num_experimentos = 1000000

somas = []

for i in range(num_experimentos):
    soma = 0
    soma += np.sum(np.random.randint(low=1,high=7, size=num_dados))
    somas.append(soma)

plt.hist(somas, bins=30, color='blue')
plt.title('Histograma de Dados Aleatórios')
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Exibindo o gráfico
plt.show()

print(np.mean(somas))
```

Ou seja, lançamento de dados que aparentemente possui distribuição uniforme, ao calcular a possibilidade de soma vira uma Normal.
