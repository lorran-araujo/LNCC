```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from sklearn.datasets import load_iris

# Carregando o conjunto de dados (Exemplo: Iris)
data = load_iris()
X = data.data

# Aplicando Kernel PCA
kpca = KernelPCA(kernel='rbf', n_components=4, fit_inverse_transform=True)
X_kpca = kpca.fit_transform(X)

# Autovalores da matriz de kernel
lambdas = kpca.eigenvalues_

# Plotando os autovalores para visualizar o critério do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, len(lambdas) + 1), lambdas, 'o-', linewidth=2)
plt.title('Autovalores do Kernel PCA')
plt.xlabel('Número de Componentes Principais')
plt.ylabel('Autovalor')
plt.grid(True)
plt.show()

# Avaliando o número ideal de componentes principais
print(f"Autovalores: {lambdas}")
```
