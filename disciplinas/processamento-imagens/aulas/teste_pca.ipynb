```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
np.random.seed(42)
```

```python
num_pontos = 1200

x = np.random.normal(0, 3, num_pontos) 
y = np.random.normal(0, 1, num_pontos)  

# Criando uma matriz de rotação
theta = np.radians(30)  # Rotação de 30 graus
R = np.array([[np.cos(theta), -np.sin(theta)], 
              [np.sin(theta), np.cos(theta)]])

# Empilhando as coordenadas em uma matriz 2xN
pontos = np.vstack((x, y))

# Aplicando a rotação
pontos_rotacionados = R @ pontos

# Separando novamente as coordenadas rotacionadas
x_rot, y_rot = pontos_rotacionados

# Plotando a nuvem de pontos elíptica rotacionada
plt.scatter(x_rot, y_rot, alpha=0.5)

# Adicionando títulos e rótulos aos eixos
plt.title("Nuvem de Pontos Elíptica Rotacionada")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# Ajustando os limites dos eixos para melhorar a visualização
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Mostrando o gráfico
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(pontos_rotacionados.T)

# Obtendo os vetores diretores (componentes principais)
componentes = 10*pca.components_  # Vetores diretores
centro = np.mean(pontos_rotacionados.T, axis=0)    # Centro dos dados

plt.figure(figsize=(6,6))

# Plotando os dados originais
plt.scatter(x_rot, y_rot, alpha=0.5)

# Plotando os vetores diretores (componentes principais)
for vetor in componentes:
    plt.quiver(centro[0], centro[1], vetor[0], vetor[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.005)

# Adicionando títulos e rótulos
plt.title('PCA: Dados e Vetores Diretores')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')

# Ajustando a escala dos eixos para ficar igual
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Mostrando o gráfico
plt.show()
```
