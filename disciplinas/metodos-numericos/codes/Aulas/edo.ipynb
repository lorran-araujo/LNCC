## Métodos de cálculo de Equações Diferenciais

# Método de Euler

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler
def euler_method(f, y0, t0, tf, h):
    t_values = np.arange(t0, tf + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
    
    return t_values, y_values
```

Exemplo:

```python
# Função que define a EDO
def f(t, y):
    return -2*(t**3)+12*(t**2)-20*t+8.5

# Parâmetros
y0 = 1  # Condição inicial
t0 = 0  # Tempo inicial
tf = 2  # Tempo final
h = 0.5  # Passo

# Cálculo usando o método de Euler
t_values, y_values = euler_method(f, y0, t0, tf, h)

for i in range(len(t_values)):
    print(f'y({t_values[i]})=({y_values[i]})')

# Plot do gráfico
plt.plot(t_values, y_values, label='Método de Euler')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Resolução da EDO usando o Método de Euler')
plt.legend()
plt.grid(True)
plt.show()
```

# Método de Heun

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Heun
def heun_method(f, y0, t0, tf, h):
    t_values = np.arange(t0, tf + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    
    for i in range(1, len(t_values)):
        y_predictor = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
        y_corrector = y_values[i-1] + h * f(t_values[i-1] + h, y_predictor)
        y_values[i] = (y_values[i-1] + y_corrector) / 2
    
    return t_values, y_values
```

Exemplo:

```python
# Função que define a EDO
def f(t, y):
    return 4*np.exp(0.8*t)-0.5*y

# Parâmetros
y0 = 2  # Condição inicial
t0 = 0  # Tempo inicial
tf = 4  # Tempo final
h = 1  # Passo

# Cálculo usando o método de Heun
t_values_heun, y_values_heun = heun_method(f, y0, t0, tf, h)

for i in range(len(t_values)):
    print(f'y({t_values_heun[i]})=({y_values_heun[i]})')

# Plot do gráfico
plt.plot(t_values_heun, y_values_heun, label='Método de Heun')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Resolução da EDO usando o Método de Heun')
plt.legend()
plt.grid(True)
plt.show()
```
