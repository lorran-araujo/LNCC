```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros iniciais
an = 2
pi = 3.141592653589793238

n = 27

# Inicializao dos arrays
I = np.zeros(n)
E = np.zeros(n)
E[0] = an - pi
I[0] = an

# Loop de iteraes
for n in range(2, n+1):
    an = (2**(n-(1/2)))*np.sqrt(1-np.sqrt(1-(4**(1-n)*(an**2))))
    I[n-1] = an
    
    erro = ((pi-an)/pi)
    E[n-1] = erro

# Plotando os grficos
iterations = np.arange(1, n+1)

plt.figure(figsize=(12, 6))

# Grfico de I
plt.subplot(1, 2, 1) #para plotar lado a lado
#plt.figure(1)
plt.plot(iterations, I, marker='o', linestyle='-', color='b')
plt.title('Valor de an em funo das Iteraes')
plt.xlabel('Iteraes')
plt.ylabel('an')
plt.grid(True)

# Grfico de E
plt.subplot(1, 2, 2)
#plt.figure(2)
plt.plot(iterations, E, marker='x', linestyle='-', color='r')
plt.title('Erro relativo em funo das Iteraes')
plt.xlabel('Iteraes')
plt.ylabel('Erro relativo')
plt.grid(True)

plt.tight_layout()
plt.show()
```
