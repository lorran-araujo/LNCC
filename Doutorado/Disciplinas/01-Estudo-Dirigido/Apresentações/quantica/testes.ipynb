Fazendo a medição de um bit quântico:

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

# 1. Criar um Circuito Quântico com 1 qubit e 1 bit clássico
# (O bit clássico serve para guardar o resultado da medição)
qc = QuantumCircuit(1, 1)

# 2. Aplicar a porta Hadamard (H) no qubit 0
# Isso tira o qubit do estado |0> e o coloca em sobreposição
qc.h(0)

# 3. Medir o qubit
# O resultado do qubit 0 será armazenado no bit clássico 0
qc.measure(0, 0)

# 4. Simular o circuito
# Usamos o 'qasm_simulator' que imita um computador quântico ideal
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1000) # Rodamos o experimento 1000 vezes
result = job.result()

# 5. Mostrar os resultados
counts = result.get_counts(qc)
print(f"Resultados das 1000 execuções: {counts}")
```

Observando o emaranhamento:

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

# 1. Criar um circuito com 2 qubits e 2 bits clássicos
qc = QuantumCircuit(2, 2)

# 2. Passo 1: Colocar o qubit 0 em sobreposição (50% de chance de ser 0 ou 1) aplicando radamart
qc.h(0)

# 3. Passo 2: Aplicar a porta CNOT
# O qubit 0 é o controle, o qubit 1 é o alvo
qc.cx(0, 1)

# 4. Medir ambos os qubits
qc.measure([0, 1], [0, 1])

# 5. Simular
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1000)
result = job.result()

# 6. Ver os resultados
counts = result.get_counts(qc)
print(f"Resultados das 1000 execuções: {counts}")
```

Observando o fenômeno da interferência:

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

# 1. Criar um circuito com 1 qubit e 1 bit clássico
qc = QuantumCircuit(1, 1)

# 2. Primeira Hadamard: Cria sobreposição 50/50
# Estado: (|0> + |1>) / sqrt(2)
qc.h(0)

# 3. Porta Z: Aplica uma fase negativa no componente |1>
# Isso não muda a probabilidade (ainda é 50/50), mas muda o SINAL (fase)
# Estado vira: (|0> - |1>) / sqrt(2)
qc.z(0)

# 4. Segunda Hadamard: Onde a INTERFERÊNCIA acontece
# Ela vai somar as amplitudes. 
# O componente |0> sofrerá interferência destrutiva ou construtiva?
qc.h(0)

# 5. Medição
qc.measure(0, 0)

# 6. Execução
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1000)
result = job.result().get_counts()

print(f"Resultado após interferência: {result}")
```
